```python
movies_df = pd.read_csv('ml-1m/movies.dat',names=['MovieID','Title','Genres'],
encoding='ISO-8859-1',sep='::',engine='python')
```

```python
movies_df
```

```python
missing =  movies_df.isnull().sum()
print("Missing values",missing)
perc = round(missing/ratings.shape[0]*100,2)
print("Percentage of Missing",perc)
```

```python
genres =   [ 'Action',
	 'Adventure',
	'Animation',
	"Children's",
	 'Comedy',
	 'Crime',
	 'Documentary',
	 'Drama',
	 'Fantasy',
	'Film-Noir',
	 'Horror',
	 'Musical',
	 'Mystery',
	 'Romance',
	 'Sci-Fi',
	 'Thriller',
	 'War',
	 'Western']
```

```python
genres
```

```python
for genre in genres:
    movies_df.loc[:,genre] = 0
```

```python
movies_df
```

```python
for row in movies_df.iterrows():
    index = row[0]
    genre_data = row[1][2]
    for genre in genres:
        if genre in genre_data:
            movies_df.loc[index, genre] = 1
    
```

```python
movies_df
```

```python
movies_df.drop(labels=['Genres'], axis=1, inplace=True)
```

```python
movies_df
```

```python
movies_df.to_csv('cleaned/movies.csv', index=False)
```
